```python
#5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
#i) Headline
#ii) Time
#iii) News Link
```

```python
page6 = requests.get('https://www.cnbc.com/world/?region=world')
```

```python
page6
```

```python
soup6 = BeautifulSoup(page6.content)
```

```python
soup6
```

```python
headline = []
for i in soup6.find_all ('div',class_="RiverPlusCard-container"):
    headline.append(i.text)
headline[ :10]
```

```python
time = []
for i in soup6.find_all ('span',class_="RiverByline-datePublished"):
    time.append(i.text)
time[0:10]
```

```python
link = []
for i in soup6.find_all ('span',class_="RiverByline-authorByline"):
    link.append(i.text)
link[0:10]
```

```python
dfnews= pd.DataFrame({"news":headline[0:10],"newstime":time[0:10],"newslink":link[0:10]})
```

```python
dfnews
```

```python
#Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
#i) Paper Title
#ii) Authors
#iii) Published Date
#iv) Paper URL
```

```python
page7 = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page7
```

```python
soup7 = BeautifulSoup(page7.content)
```

```python
soup7
```

```python
papertitle = []
for i in soup7.find_all ('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    papertitle.append(i.text)
papertitle
```

```python
authors = []
for i in soup7.find_all ('span',class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)
authors
```

```python
date = []
for i in soup7.find_all ('span',class_="sc-1thf9ly-2 dvggWt"):
    date.append(i.text)
date
```

```python
link = []
for i in soup7.find_all ('a',class_="sc-5smygv-0 fIXTHm"):
    link.append(i.text)
link
```

```python
df7 = pd.DataFrame({"Title":papertitle,"Author":authors,"Publishdate":date,"Weblink":link})
```

```python
df7
```

```python
#Write a python program to scrape mentioned details from dineout.co.in and make data frame-
#i) Restaurant name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image URL
```

```python
page8 = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page8
```

```python
soup8 = BeautifulSoup(page8.content)
```

```python
soup8
```

```python
rname = []
for i in soup8.find_all ('a',class_="restnt-name ellipsis"):
    rname.append(i.text)
rname
```

```python
style = []
for i in soup8.find_all ('span',class_="double-line-ellipsis"):
    style.append(i.text)
style
```

```python
location = []
for i in soup8.find_all ('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
rating = []
for i in soup8.find_all ('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
```

```python
class="no-img"
```

```python
url = []
for i in soup8.find_all ('img',class_="no-img"):
    url.append(i['data-src'])
url
```

```python
df8 = pd.DataFrame({"Restaurant name":rname,"Cuisine":style,"Place":location,"Restaurant Rating":rating,"Image url":url})
```
